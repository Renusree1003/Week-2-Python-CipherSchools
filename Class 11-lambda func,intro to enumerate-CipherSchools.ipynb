```python
names = ("jatin", "saransh", "shubhan", "samarth")
scores = (50, 80, 90, 70)
for i, name in zip(names):
    score = scores(i)
    print(name, "_", score)
```

```python
names = ("jatin", "saransh", "shubham", "samarth")
scores = (50, 80, 90, 70)
states = ("delhi", "punjab", "punjab", "punjab")
for name, score, state in zip(names, scores, states):
    print(name, "_", score, state)
```
