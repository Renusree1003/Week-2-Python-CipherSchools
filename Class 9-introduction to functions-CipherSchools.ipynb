```python
def show_loading():
    for _ in range(3):
        print("waiting...")

a = 5
b = 7
print(a + b)
show_loading()
print(a-b)
show_loading()
print(a*b)
show_loading()
```

```python
def sheldon_knock(name):
    for _ in range(3):
        print("knock knock knock", name)
sheldon_knock("leonard")
```
